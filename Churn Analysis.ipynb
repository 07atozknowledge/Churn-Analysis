```python
import pandas as pd 
import numpy as  np
import matplotlib.pyplot as plt
import seaborn as sns 
```

## Fetching Data


```python
df = pd.read_csv('Customer Churn.csv')
df.info()
```

## Data Cleaning and Processing

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
df.info()

# Total charges changed from object to float data types.
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
def convo(value):
    if value == 1:
        return "Yes"
    else:
        return "No"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(convo)

# Conversion of SeniorCitizen 0 and 1 to Yes and No text format.
```

```python
df.head(30)
```

```python
cx = sns.countplot( x = 'Churn' , data = df)
cx.bar_label(cx.containers[0])
plt.title("Count of Customer by Churn",fontsize = 10)
plt.show()
```

```python
plt.figure(figsize=(3,4))
group_by = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(group_by['Churn'],labels = group_by.index ,autopct = "%1.2f%%")
plt.title("Percentage of Churned Customer",fontsize = 10)
plt.show()
group_by
```

```python
# From the above pie chart it is clear that out of total , 26.54% customer have churned out.
```

```python
plt.figure(figsize=(4,4))
gx = sns.countplot( x = 'gender' , data = df , hue = 'Churn')
gx.bar_label(gx.containers[0])
plt.title("Count of Customer by Gender",fontsize = 10)
plt.show()
```

```python
# Churn rate of both genders are almost equal
```

```python
plt.figure(figsize=(4,4))
gx = sns.countplot( x = 'SeniorCitizen' , data = df , hue = 'Churn')
gx.bar_label(gx.containers[0])
plt.title("Count of Customer by SeniorCitizen",fontsize = 10)
plt.show()
```

```python
tot_cot = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack()*100

fig, ax = plt.subplots(figsize=(6,6))

tot_cot.plot(kind='bar',stacked=True,ax=ax,color=['#1f77b4','#ff7f0e'])

for p in ax.patches:
    width , height  = p.get_width(),p.get_height()
    x,y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha = 'center' , va = 'center')


plt.title('Churn by Senior Citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage(%)')
plt.xticks(rotation=0)
plt.legend(title='Churn',loc='upper right')

plt.show()
```

```python
# rate of Senior Citizen Churn is  more
```

```python
# Columns you want to plot
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection','TechSupport', 'StreamingTV', 'StreamingMovies']

# Setup figure and axes
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))
axes = axes.flatten()  # Flatten to 1D array for easy looping

# Loop through columns and axes
for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, ax=axes[i], hue=df["Churn"])
    axes[i].set_title(f'Count of {col}', fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=0)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,5))  # Adjusted for space
gx = sns.countplot(x='PaymentMethod', data=df, hue='Churn')
gx.bar_label(gx.containers[0])
gx.bar_label(gx.containers[1])  # Add labels for second hue as well

# Move legend outside the plot (to the right)
plt.legend(title='Churn', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title("Count of Customer by Payment Method", fontsize=10)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# More churn in electronic pay methods
```
